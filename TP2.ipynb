```python
#Question 3

v=R.groupby("variety").price.unique()
v.index
```

```python
Table_max=[]
for i in range (len(v.index)):
    M=max(v[i])
    Table_max.append(M)
Table_max
```

```python
Table_min=[]
for i in range (len(v.index)):
    M=min(v[i])
    Table_min.append(M)
Table_min
```

```python
d={"Max": Table_max, "min": Table_min}
d1=pd.DataFrame(d, index=v.index)
d1
```
